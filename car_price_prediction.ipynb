The data set we will be working with contains information on various cars. For each car we have information about the technical aspects of the vehicle such as the motor's displacement, the weight of the car, the miles per gallon, how fast the car accelerates, and more.

### Data Cleaning

```python
# Select only the columns with continuous values
conti_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_data = data[conti_cols]
```

```python
numeric_data.head()
```

```python
numeric_data = numeric_data.replace("?",np.nan)
```

```python
numeric_data.isnull().sum()
```

```python
numeric_data = numeric_data.astype('float')
```

```python
numeric_data.dtypes
```

#### Handling missing values

```python
numeric_data.isnull().sum()
```

#### It shows that normalized losses column contains maximum missing values and bore, stroke, horsepower, peak-rpm,price are columns having missing values

```python
# Because `price` is the column we want to predict, let's remove any rows with missing `price` values.

numeric_data = numeric_data.dropna(subset=['price'])
numeric_data.isnull().sum()
```

```python
## Replace missing values in other columns using column means.
numeric_data = numeric_data.fillna(numeric_data.mean())
```

```python
numeric_data.isnull().sum()
```

#### It's clear now that their is no more missing values in our dataset

### Normalization

```python
# rescaling all the training columns from 0 to 1

numeric_data = (numeric_data - numeric_data.mean()) / numeric_data.std()
```
